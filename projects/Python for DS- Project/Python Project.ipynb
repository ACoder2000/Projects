# Task 1: Number of characters

```python
len(characters)
```

Hence the number of characters in the given script were 17

```python
dialogues={}
for d in characters:
    dialogues[d]=''
```

```python
for i in chardiag:
    for c in characters:
        if(i[0]==c):
            dialogues[c]=dialogues[c]+' '+i[1]
            break
```

```python
def myfun(string):
    cleanedwords=set()
    words=string.split(' ')
    for i in words:
        cleanedwords.add(''.join(filter(str.isalnum, i)))
    return cleanedwords
```

```python
for c in characters:
    file=open('untitled folder/'+c+'.txt','w+')
    cleanstr=myfun(dialogues[c].lower())
    cleanstr.remove('')
    file.seek(0)
    for i in cleanstr:
        file.write(i+'\n')
    file.close()
```
